```python
def getStockPrice(ticker):
    url = f'https://finance.naver.com/item/main.naver?code={ticker}'
    res = requests.get(url)
    html = bts(markup = res.text, features = 'html.parser')
    return html.select('p.no_today > em > span.blind')[0].text.replace(',', '')
```

```python
getStockPrice(ticker = '136490')
```
